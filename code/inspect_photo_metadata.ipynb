```python
import os 
import pandas as pd
from glob import glob as gb
import matplotlib.pyplot as plt
```

```python
list_imgs = gb('/media/ruben/OSDisk/Users/ruben.ros/Documents/GitHub/VisualTopicModelling/data/photos/*')
```

```python
sizes = [os.path.getsize(path) / 1000 for path in list_imgs]
```

```python
plt.hist(sizes, bins=2048)
plt.xlim(0,1024)
plt.title('Hist. of Image Sizes')
plt.ylabel('Number of Images')
plt.xlabel('Size in Kilobytes (Kb) (0-1024)')
plt.show()
```
